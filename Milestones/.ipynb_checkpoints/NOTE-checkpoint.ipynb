```python
import pandas as pd
import numpy as np
import scipy.stats as stat
import matplotlib.pyplot as plt
import seaborn as sb
from pygal import *
from IPython.display import SVG, HTML
sb.set()
```

```python
# Boston datasets
listing = "../Airbnb-Datasets/Boston/listings.csv"
review = "../Airbnb-Datasets/Boston/reviews.csv"
boston_listings = pd.read_csv(listing)
boston_reviews = pd.read_csv(review)

# Washington datasets
listing = "../Airbnb-Datasets/Washington/Listings - Detailed.csv"
review = "../Airbnb-Datasets/Washington/Reviews - Detailed.csv"
was_listings = pd.read_csv(listing)
was_reviews = pd.read_csv(review)

# New York datasets
listing_kaggle = "../Airbnb-Datasets/New York/AB_NYC_2019.csv"
listing = "../Airbnb-Datasets/New York/airbnb-listings.csv"
ny_listing_kaggle = pd.read_csv(listing_kaggle)
ny_listings = pd.read_csv(listing, sep=';')

# San Francisco datasets
listing = "../Airbnb-Datasets/San Francisco/San francisco.csv"
san_listings = pd.read_csv(listing)
```

```python
# Number of hosts mentioned in the table
boston_listings['host_id'].nunique()
# Number of locations mentioned in the table

neighbor = boston_listings['neighbourhood'].unique()
price_by_neighbor = {i:[] for i in neighbor}
for each in neighbor:
    data = boston_listings[boston_listings['neighbourhood'] == each]['price']
    data = pd.to_numeric(data.apply(lambda x: x.strip("$")),errors='coerce').dropna()
    price_by_neighbor[each] = data.values

plt.figure(figsize=(15,6), dpi=100)
plt.boxplot(price_by_neighbor.values())
plt.xticks(np.arange(1,32), price_by_neighbor.keys(), rotation=75);
```

```python
# Find the correlation between price and square_feet
price = boston_listings['price'].apply(lambda x: x.strip("$").replace(",","")).astype(float).values
rooms = boston_listings[['bathrooms','bedrooms']].mean(axis=1).values

slope, intercept, r_value, p_value, std_err = stat.linregress(rooms, price)
predicted_price = slope * rooms + intercept

plt.figure(figsize=(10,6), dpi=100)
plt.scatter(rooms, price)
plt.plot(rooms, predicted_price);
plt.xlabel("Average number of rooms")
plt.ylabel("Price")
plt.title("The relationship between average number of rooms and price");

text = "Slope: {} - Intercept: {} - p_value: {} - std_err: {}".format(round(slope), round(intercept), round(p_value), round(std_err))
plt.text(x=1.2,y=4450,s='50',text=text);
```
